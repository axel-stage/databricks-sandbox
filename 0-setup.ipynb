## Project Setup
1. create catalog: sandbox_catalog
2. create schemas: landing, bronze
3. create volume in landing schema
4. delete default schema

```sql
show catalogs;
```

```sql
create catalog if not exists sandbox_catalog
  managed location 's3://lakehouse-sandbox-qzoj/'
  comment 'sandbox location for testing';
```

```sql
use catalog sandbox_catalog;
select current_catalog();
```

```sql
create schema if not exists landing
  managed location 's3://lakehouse-sandbox-qzoj/landing';

create schema if not exists bronze
  managed location 's3://lakehouse-sandbox-qzoj/bronze';
```

```sql
create external volume if not exists landing.data
  location 's3://lakehouse-sandbox-qzoj/landing/data';
```

```sql
drop schema if exists default;
```
